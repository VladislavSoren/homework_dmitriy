```python
class FileReader:

    file_readers_count = 0  # атрибут класса

    def __init__(self, filename):  # FileReader("file.txt")
        self.filename = filename  # атрибут объекта класса
        self.file = None
        FileReader.file_readers_count += 1

    def __iter__(self):  # iter(file_reader)
        self.file = open(self.filename)
        return self.file
    
    @classmethod
    def decrese_readers_count_on(cls, num):  # метод класса
        cls.file_readers_count -= num

    @staticmethod  # статический метод, т.е. НЕ содержит ссылки на объекты и классы
    def clear_line(line: str) -> str:
        clear_line = line.replace("#", "").replace(" ", "").replace("\n", "")
        return clear_line

    def __next__(self):  # метод объекта класса, потомучто self - ссылка на объект
        line = self.file.readline()
        line = self.clear_line(line)  # вызо стат. метода clear_line

        if line == "<end>":
            self.file.close()
            raise StopIteration
        else:
            return line
```
