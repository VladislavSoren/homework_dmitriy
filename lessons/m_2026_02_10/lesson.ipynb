```python
# локальная переменная это та, которую мы передаём в функцию (как параметры) или объявляем внутри
# в противном случае это enclosing переменная. enclosing переменные нужно искать, локальные и так видны.

def timer(timeout): # timeout - local для timer, timeout - enclosing для wrapper_out, timeout - enclosing для wrapper
    
    def wrapper_out(fun): # fun - local для wrapper_out, fun - enclosing для функции wrapper
        
        def wrapper(name, age):
            print(f"timeout - {timeout}") # timeout - enclosing для wrapper
            fun(name, age)         # name, age - local для wrapper, fun - enclosing для wrapper
            print(123)
        return wrapper
    
    return wrapper_out


# @timer(5)
def fun(name, age=123, sex='m'): 
    print(name, age)
    return name, age

wrapper_out = timer(5)
fun = wrapper_out(fun) # fun = wrapper
   
fun("Bob", 25, "w")

# timeout
```

```python
# локальная переменная это та, которую мы передаём в функцию (как параметры) или объявляем внутри
# в противном случае это enclosing переменная. enclosing переменные нужно искать, локальные и так видны.

def timer(timeout): # timeout - local для timer, timeout - enclosing для wrapper_out, timeout - enclosing для wrapper
    
    def wrapper_out(fun): # fun - local для wrapper_out, fun - enclosing для функции wrapper
        
        def wrapper(*args): # args - кортеж с позиционными аргументами
            print(f"timeout - {timeout}") # timeout - enclosing для wrapper
            fun(*args)         # name, age - local для wrapper, fun - enclosing для wrapper
            print(123)
        return wrapper
    
    return wrapper_out


@timer(5)
def fun(name, age=123, sex='m'): 
    print(name, age)
    return name, age

# wrapper_out = timer(5)
# fun = wrapper_out(fun) # fun = wrapper
   
fun(5)
```
